#A Frequency Spectrum of Somali Letters

```python
# import libraries
import matplotlib.pyplot as plt
import re
from tabulate import tabulate
```

```python
# read data
with open('data/text/somali-phrases-corpus.txt', 'r') as f:
    text = f.read()
```

```python
# count frequency of each character
cleaned_text = re.sub(r"[^a-zA-Z]", "", text)
counts = {}
for char in cleaned_text.lower():
    if char.isalpha():
        counts[char] = counts.get(char, 0) + 1
```

```python
# calculate percentages
total = sum(counts.values())
percentages = {}

for k, v in counts.items():
    percentage = round((v/total) * 100, 2)
    percentages[k] = percentage
```

```python
# Show data as a table
table = tabulate({"Alphabet": percentages.keys(
), "Percentage %": percentages.values()}, headers="keys", tablefmt='html')
table
```

```python
# show frequency on a bar chart
plt.bar(percentages.keys(), percentages.values(), color='blue', alpha=0.6)
for i, value in enumerate(percentages.values()):
    plt.text(i, value, str(value), ha='center')
plt.title('Somali Alphabet Frequency')
plt.xlabel('Alphabet')
plt.ylabel('%')
plt.show()
```
